```python
import cv2 as cv 
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

### Leitura da imagem e segmentação dos canais da imagem (RGB)

```python
img = cv.imread('colorida.jpg')
azul, verde, vermelho = cv.split(img)
```

### Mostrar a Imagem

```python
plt.imshow(img)
```

### Histograma da Imagem

```python
fig = plt.figure(figsize=(20,5))

ax1 = fig.add_subplot(131) # 1 linha 3 colunas 1 referene a primeira imagem
ax1.hist(azul.ravel(), 256, [0,256]) # a função hist é para mostrar o histograma da cor
plt.title('Histograma do canal azul')

ax2 = fig.add_subplot(132) # 1 linha 3 colunas 2 referene a segunda imagem
ax2.hist(verde.ravel(), 256, [0,256])
plt.title('Histograma do canal verde')

ax3 = fig.add_subplot(133) # 1 linha 3 colunas 3 referene a terceira imagem
ax3.hist(vermelho.ravel(), 256, [0,256])
plt.title('Histograma do canal vermelho')

plt.show()


```
