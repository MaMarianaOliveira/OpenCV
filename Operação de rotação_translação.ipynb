```python
import cv2 as cv
import numpy as np
from matplotlib import pyplot as plt
```

### Lendo a imagem

```python
imgOriginal = cv.imread('recife.jpg',0)
```

```python
totalLinhas, totalColunas = imgOriginal.shape
print(totalLinhas)
print(totalColunas)
```

### Rotacionando a imagem

```python
mat = cv.getRotationMatrix2D((totalLinhas/2, totalColunas/2), 45, 1)
```

```python
imgRotacionada = cv.warpAffine(imgOriginal, mat, (totalColunas,totalLinhas))
```

```python
fig = plt.figure(figsize=(10,8))
plt.imshow(imgRotacionada, cmap = plt.cm.gray)
```

### Translao

```python
imgOriginal = cv.imread('recife.jpg',0)
```

```python
matriz =np.float32([[1,0,100],[0,1,100]])
imgDeslocada = cv.warpAffine(imgOriginal, matriz,(totalColunas,totalLinhas))
```

```python
fig = plt.figure(figsize=(10,8))
plt.imshow(imgDeslocada, cmap = 'gray')
```

### Reajuste de Escala da Imagem

```python
imgModificada = cv.resize(imgOriginal,
                         None,
                         fx=0.8,
                         fy=0.8,
                         interpolation = cv.INTER_CUBIC)
```

```python
cv.imshow('Imagem Original', imgOriginal)
cv.imshow('Imagem Modificada', imgModificada)
cv.waitKey(0)
cv.destroyAllWindows()

```
