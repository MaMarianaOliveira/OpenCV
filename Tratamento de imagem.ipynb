```python
import cv2 
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('recife.jpg', 0)
```

```python
imagemEqualizada = cv2.equalizeHist(img)
```

```python
fig = plt.figure(figsize = (20,5)) #Imagem recebida
ax1= fig.add_subplot(121)
plt.imshow(img , cmap = plt.cm.gray)


ax1= fig.add_subplot(122) #Imagem tratada
plt.imshow(imagemEqualizada, cmap= plt.cm.gray)


plt.show()
```

```python
fig = plt.figure(figsize = (20,5)) #Imagem recebida
ax1= fig.add_subplot(121)
plt.hist(img.ravel(), 256, [0, 256])


ax1= fig.add_subplot(122) #Imagem tratada
plt.hist(imagemEqualizada.ravel(), 256, [0, 256])


plt.show()
```
